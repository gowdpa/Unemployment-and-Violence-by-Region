Often you will want to select certain rows, columns, or specific cells from a large dataframe. This can be accomplished using indexing. There are some limitations, however.

For example, we can use numbers in square brackets to select certain rows, but doing so always returns all the columns in our dataset, such as the function: `df [0:3]`

If we use the `.iloc` attribute before our brackets, pandas accepts two numbers separated by a comma. The first number is for rows and the second for columns. This will give us the information of a specefic cell, as shown in the following function:


```python
df.iloc[2,3]
```

We can also select information for specific columns using the column names, however, you MUST type out the column name, and you cannot use the column number.

The dataset can also be filtered by using logical conditions (based on true or false), these can be added using nested square brackets.

See the example below.

- The inner statement, `df["State"]=="RI"` looks for a column name and checks if it equals `"RI"`
- The outer statement `df[ ... ]` uses the resulting column of `True/False` values to select rows
- When combined, these two commands call all of the data contained in rows where the value of the `State` field is equal to `"RI"`

```python
df[df["State"] == "RI"]
```

To select multiple columns, it must be listed in its own nested set of square brackets.

See the examples below

```python
df[["State","Region","Division","County","Unemployment"]]
```

```python
df[["State","Region","Division","County","Unemployment","Violent crime"]]
```

`.iloc` is used for intergers, whereas `.loc` is used for column names.

To index with loc, the function `df.loc[:,[...]]` is used. If specific rows were selected, those numbers would go where the semi colon is: `df.loc[0:3,[...]]`. The values inside of the second brackets are the column names. To recieve a random sample, `.sample(n=)` is added at the end of the function.

All together, the function should appear like this: `df.[:,[...]].sample(n=)`



```python
df.loc[:,["State","Region","Division","County","Unemployment","Violent crime"]].sample(n=50)
```

```python
df.loc[:,["State","Region","Division","Unemployment","Violent crime"]].sample(n=50)
```

```python
df.loc[:,["State","Region","Unemployment","Violent crime"]].sample(n=300)
```

####Creating a New Dataframe
Now a new dataframe needs to be created for this data set. This will be useful to export this data.

To do this, name the data using the same process as initally, but instead  of using `pd.read_csv()`we will use the function that created our data set.

Set the name of your new dataframe (here I am using `UseableData`) equal to the function that created your data set.

```python
UseableData=df.loc[:,["State","Region","Unemployment","Violent crime"]].sample(n=300).copy()
```

We can run `UseableData` to test that we did the previous step correctly.

```python
UseableData
```
